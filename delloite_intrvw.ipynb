```python
'''
Merge Two Dictionaries:
 Combine dict1 = {"a": 1, "b": 2} and dict2 = {"b": 3, "c": 4}. 
 If a key exists in both, sum their values.
'''

f1={"a": 1, "b": 2} 
f2={"b": 3, "c": 4}
res=f1.copy()
for keys,values in f2.items():
    if keys in res:
        res[keys]+=values
    else:
        res[keys]=values

print(res)

from collections import Counter
merg=dict(Counter(f1)+Counter(f2))
print(merg)


    
```

```python
'''
Given a list of integers, return all elements that appear more than once.
'''
arr=[1,1,2,3,4,4,5,6,5]
# arr=[1,1]
d=dict()
res=[]
for i in arr:
    if i in d:
        d[i]+=1
    else:
        d[i]=0
print(d)
for key,value in d.items():
    if value>=1:
        res.append(key)
print(res)
```

```python
''' 
Second Largest Number:
 Find the second largest number in a list without using the built-in sort()
'''

arr=[7,12,5,6,3,11]
sec=first=-1
for i in arr:
    if first<i :
        sec=first
        first=i
    if sec<i and first>i:
        sec=i

print(sec,first)
```

```python
'''
pandas
'''

import pandas as pd

# Raw Data
sales_data = [
    {'Product': 'Laptop', 'Category': 'Electronics', 'Revenue': 1200},
    {'Product': 'Mouse', 'Category': 'Electronics', 'Revenue': 50},
    {'Product': 'Desk', 'Category': 'Furniture', 'Revenue': 300},
    {'Product': 'Laptop', 'Category': 'Electronics', 'Revenue': 1100},
    {'Product': 'Chair', 'Category': 'Furniture', 'Revenue': 150},
    {'Product': 'Mouse', 'Category': 'Electronics', 'Revenue': 45}
]

df=pd.DataFrame(sales_data)
df.head()
```

```python
df.head()
'''
Find the Total Revenue for each Category.
'''
tot_rev=df.groupby('Category')['Revenue'].sum()
print(tot_rev)
```

```python
'''
Find the Average Revenue per Product.
'''
avg_rev_per_prod=df.groupby('Product')['Revenue'].mean()
avg_rev_per_prod
```

```python
# Table 1: Sales
df_sales = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Desk'],
    'Revenue': [1200, 50, 300]
})

# Table 2: Suppliers
df_suppliers = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Monitor'],
    'Supplier': ['Dell', 'Logitech', 'Samsung']
})

res=pd.concat([df_sales,df_suppliers],axis=0)
print(res)
left_join_data=pd.merge(df_sales,df_suppliers,on='Product' , how='left')
print(left_join_data)
```

```python
res=left_join_data.fillna('unknown')
res['Profit']=0
res['Profit']=res['Profit'].fillna(0).astype(float)
res['Profit']=0.15*res['Revenue']
```

```python
class BankAccount:
    def __init__(self, balance):
        # PRIVATE variable (Hidden from the outside)
        self.__balance = balance 

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance is {self.__balance}")

    def get_balance(self):
        # Public way to "peek" at the private data
        return self.__balance

# Create an account
my_acc = BankAccount(1000)

# 1. This will FAIL (AttributeError)
# print(my_acc.__balance) 

# 2. You MUST use the methods provided
my_acc.deposit(500)
print(f"Current Balance: {my_acc.get_balance()}")
```
